```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def crosscorrelation(signal1,signal2):
    crosscorr=np.correlate(signal1,signal2,mode='full')
    return crosscorr
```

```python
def autocorrelation(signal):
    autocorr=np.correlate(signal,signal,mode='full')
    return autocorr
```

```python
signal1 = np.array([1, 2, 3, 4, 5])
signal2 = np.array([2, 4, 6, 8, 10])
```

```python
crosscorr=crosscorrelation(signal1,signal2)
```

```python
auto_corr = autocorrelation(signal1)
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.stem(crosscorr)
plt.title('Cross-correlation')
plt.xlabel('Time Lag')
plt.ylabel('Magnitude')
plt.subplot(2, 1, 2)
plt.stem(auto_corr)
plt.title('Autocorrelation')
plt.xlabel('Time Lag')
plt.ylabel('Magnitude')
plt.tight_layout()
plt.show()
```
